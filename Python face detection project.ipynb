```python
import cv2
```

#haarcascade_frontalface_default.xml

```python
cascade_face = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
capture=cv2.VideoCapture(0)
```

```python
while True:
    ret, img = capture.read()
    print(ret)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)         #converting bgr into grayscale image
    face = cascade_face.detectMultiScale(gray,  1.3, 8)
    
    for(x, y, w, h) in face:
        cv2.rectangle(img, (x,y), (x + w, y + h), (0,0,255), 8)
        
    cv2.imshow('img', img)
    k=cv2.waitKey(30) & 0xff
    if k == 27:
        break
        
capture.release()
cv2.destroyAllWindows()
```
